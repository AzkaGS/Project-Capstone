```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = "/content/ObesityDataSet.csv"
data = pd.read_csv(url)
```

```python
# 1. Tampilkan beberapa baris pertama dari dataset
print("Beberapa baris pertama dari dataset:")
print(data.head())
```

```python
print(data.info())
```

```python
# Deskripsi statistik dataset
print("\nDeskripsi statistik dari dataset:")
print(data.describe())
```

```python
# Visualisasi distribusi tingkat obesitas
plt.figure(figsize=(10, 6))
sns.countplot(data['NObeyesdad'])
plt.title('Distribusi Tingkat Obesitas')
plt.xticks(rotation=45)
plt.xlabel('Tingkat Obesitas')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Cek missing values
print("\nCek keberadaan missing values:")
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# 6. Cek data duplikat
print("\nJumlah data duplikat:")
print(data.duplicated().sum())
```

```python
# 7. Cek keseimbangan data
print("\nKeseimbangan data untuk variabel target:")
print(data['NObeyesdad'].value_counts())
```

```python
# Menggunakan melt untuk mengubah data menjadi format panjang
melted_data = data[['Height', 'Weight', 'Age']].melt(var_name='Feature', value_name='Value')

# Deteksi outlier menggunakan boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Feature', y='Value', data=melted_data)
plt.title('Deteksi Outlier Menggunakan Boxplot')
plt.ylabel('Nilai')
plt.show()
```

# Kesimpulan
## Dataset memiliki beberapa missing values yang perlu ditangani, terdapat data duplikat yang perlu dihapus, keseimbangan data menunjukkan bahwa beberapa kategori mungkin lebih dominan, terdapat outlier yang terdeteksi pada fitur tinggi badan dan berat badan.
